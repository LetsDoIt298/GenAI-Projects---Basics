```python
# Read the HTML file from your system
with open('C:/Users/alank/Downloads/Best Universities in Germany for International Students.html', 'r', encoding='utf-8') as file:
    content = file.read()
soup = BeautifulSoup(content, 'html.parser')
```

```python

#University Name
uni=[]
temp = soup.find_all('h3', class_="jsx-3040859135 mb-0 d-inline")
for i in range(len(temp)):
    uni.append(temp[i].get_text())
```

```python
country=[]
state=[]
acceptance_rate=[]
temp = soup.find_all('span', class_="jsx-3040859135 d-flex align-items-start justify-content-start")
for i in range(len(temp)):
    st=temp[i].get_text(' ').split(' ')
    if len(st) >5 :        
        state.append(st[len(st)-len(st)]+' '+st[len(st)-5][:-1])
        country.append(st[len(st)-4])
        acceptance_rate.append(st[len(st)-3])
    else:
        state.append(st[len(st)-len(st)][:-1])
        country.append(st[len(st)-4])
        acceptance_rate.append(st[len(st)-3])
```

```python
temp[0].get_text()
```

```python
tusion=[]
temp = soup.find_all('span', class_="jsx-3040859135 font-weight-bold text-success")
for i in range(len(temp)):
    tusion.append(temp[i].get_text().split('/')[0])
```

```python
living=[]
temp = soup.find_all('span', class_="jsx-3040859135 font-weight-bold text-success")
for i in range(len(temp)):
    living.append(temp[i].get_text().split('/')[0])

course=[]
temp = soup.find_all('span', attrs={'title': True})
for i in range(len(temp)):
    course.append(temp[i]['title'])
```

```python
soup.find_all('span', class_="jsx-3040859135 font-weight-bold text-success")
```

```python
pd.DataFrame({'Country':country, 'State':state, 'University':uni, 'Course Name':course,
              'Acceptance Rate':acceptance_rate, 'Fees':tusion[50],'Living Cost':living[50]}).to_excel('1st.xlsx')
```
